# TP 1: Algos de Tris + Complexité

### Les Algorithmes de Tris

```python
import random
import time
import timeit 

N=100
list1= [None]*N
def reinit():
    global list1
    for i in range(N):
        list1[i]=(random.randint(1,1546612))
reinit()
print(list1)
```

1. **Tri par insertion**


```python
def trierInsertion(list1,N):

    for i in range(1,N):
        temp=list1[i]
        j=i-1
        while(j>=0 and temp<list1[j]):
            list1[j+1]=list1[j]
            j-=1

        list1[j+1]=temp

print(list1[12])
```

**- Calcul de Complexité**

```python
moyenT=0

for k in range(10): #from 0 to 9
    reinit()

    debut = time.time()
    trierInsertion(list1,N)
    fin = time.time()

    moyenT+=fin-debut

print(f"Le temps moyen quand N={N} par insertion est: {moyenT/10}")
```

2. **Tri par selection**

```python
def trierSelection(list1,N):
    for i in range(N-1):
        min_indice=i
        for j in range(i+1,N):
            if list1[min_indice]>list1[j]:
                min_indice=j
        tempsel=list1[i]
        list1[i]=list1[min_indice]
        list1[min_indice]=tempsel

trierSelection(list1,N)
print(list1)
```

**- Calcul de Complexité**

```python
moyenT=0

for k in range(10): #from 0 to 9
    reinit()

    debut = time.time()
    trierSelection(list1,N)
    fin = time.time()
    
    moyenT+=fin-debut

print(f"Le temps moyen quand N={N} par selection est: {moyenT/10}")
```

3. **Tri par bulles**

```python
def trierBulles(list1,N):
    for i in range(N-1):
        for j in range(N-(i+1)):
            if list1[j]>list1[j+1]:
                tempsel=list1[j]
                list1[j]=list1[j+1]
                list1[j+1]=tempsel

trierBulles(list1,N)
print(list1)
```

**- Calcul de Complexité**

```python
moyenT=0

for k in range(10): #from 0 to 9
    reinit()

    debut = time.time()
    trierBulles(list1,N)
    fin = time.time()
    
    moyenT+=fin-debut

print(f"Le temps moyen quand N={N} par bulles est: {moyenT/10}")
```

4. **Recherche dichotomique**

```python
def rechdicho(list1,N,val):
    debut = 0
    fin = N - 1

    while debut <= fin:
        milieu = (debut + fin) // 2

        if list1[milieu] == val:
            return milieu  
        elif val < list1[milieu]:
            fin = milieu - 1
        else:
            debut = milieu + 1

    return -1
    
print(rechdicho(list1,N, list1[15]))
```

**- Calcul de Complexité**

```python
moyenT=0

for k in range(10): #from 0 to 9
    reinit()

    debut = time.time()
    rechdicho(list1,N, list1[15])
    fin = time.time()
    
    moyenT+=fin-debut

print(f"Le temps moyen quand N={N} recherche dicho est: {moyenT/10}")
```

### Vérification de la Complexité - Courbes de Comparaison

```python
import matplotlib.pyplot as plt
import numpy as np

# Différentes valeurs de N à tester
tailles = [10, 50, 100, 200, 500, 1000, 2000]

# Stocker les temps moyens pour chaque algorithme
temps_insertion = []
temps_selection = []
temps_bulles = []
temps_dicho = []
temps_fusion = []

# Mesurer le temps pour chaque taille
for n in tailles:
    print(f"Testing N = {n}...")
    
    # Tri par insertion
    moyenT = 0
    for k in range(10):
        list_temp = [random.randint(1, 1546612) for _ in range(n)]
        debut = time.time()
        trierInsertion(list_temp, n)
        fin = time.time()
        moyenT += fin - debut
    temps_insertion.append(moyenT / 10)
    
    # Tri par sélection
    moyenT = 0
    for k in range(10):
        list_temp = [random.randint(1, 1546612) for _ in range(n)]
        debut = time.time()
        trierSelection(list_temp, n)
        fin = time.time()
        moyenT += fin - debut
    temps_selection.append(moyenT / 10)
    
    # Tri par bulles
    moyenT = 0
    for k in range(10):
        list_temp = [random.randint(1, 1546612) for _ in range(n)]
        debut = time.time()
        trierBulles(list_temp, n)
        fin = time.time()
        moyenT += fin - debut
    temps_bulles.append(moyenT / 10)
    
    # Tri par fusion
    moyenT = 0
    for k in range(10):
        list_temp = [random.randint(1, 1546612) for _ in range(n)]
        debut = time.time()
        tri_par_fusion(list_temp.copy())
        fin = time.time()
        moyenT += fin - debut
    temps_fusion.append(moyenT / 10)
    
    # Recherche dichotomique (sur liste triée)
    moyenT = 0
    for k in range(10):
        list_temp = sorted([random.randint(1, 1546612) for _ in range(n)])
        val = list_temp[n // 2]
        debut = time.time()
        rechdicho(list_temp, n, val)
        fin = time.time()
        moyenT += fin - debut
    temps_dicho.append(moyenT / 10)

print("Mesures terminées!")
```

```python
# Créer les graphiques de vérification de complexité
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Convertir en arrays numpy
tailles_np = np.array(tailles)
temps_insertion_np = np.array(temps_insertion)
temps_selection_np = np.array(temps_selection)
temps_bulles_np = np.array(temps_bulles)
temps_dicho_np = np.array(temps_dicho)
temps_fusion_np = np.array(temps_fusion)

# Graphique 1: T(n) / n
ax1 = axes[0, 0]
ax1.plot(tailles_np, temps_insertion_np / tailles_np, 'o-', label='Insertion', linewidth=2)
ax1.plot(tailles_np, temps_selection_np / tailles_np, 's-', label='Sélection', linewidth=2)
ax1.plot(tailles_np, temps_bulles_np / tailles_np, '^-', label='Bulles', linewidth=2)
ax1.plot(tailles_np, temps_fusion_np / tailles_np, 'd-', label='Fusion', linewidth=2)
ax1.set_xlabel('Taille N', fontsize=12)
ax1.set_ylabel('Temps / N (secondes)', fontsize=12)
ax1.set_title('Complexité O(n) - Temps divisé par N', fontsize=14, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Graphique 2: T(n) / n²
ax2 = axes[0, 1]
ax2.plot(tailles_np, temps_insertion_np / (tailles_np ** 2), 'o-', label='Insertion', linewidth=2)
ax2.plot(tailles_np, temps_selection_np / (tailles_np ** 2), 's-', label='Sélection', linewidth=2)
ax2.plot(tailles_np, temps_bulles_np / (tailles_np ** 2), '^-', label='Bulles', linewidth=2)
ax2.plot(tailles_np, temps_fusion_np / (tailles_np ** 2), 'd-', label='Fusion', linewidth=2)
ax2.set_xlabel('Taille N', fontsize=12)
ax2.set_ylabel('Temps / N² (secondes)', fontsize=12)
ax2.set_title('Complexité O(n²) - Temps divisé par N²', fontsize=14, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Graphique 3: T(n) / (n log n)
ax3 = axes[1, 0]
nlogn = tailles_np * np.log2(tailles_np)
ax3.plot(tailles_np, temps_insertion_np / nlogn, 'o-', label='Insertion', linewidth=2)
ax3.plot(tailles_np, temps_selection_np / nlogn, 's-', label='Sélection', linewidth=2)
ax3.plot(tailles_np, temps_bulles_np / nlogn, '^-', label='Bulles', linewidth=2)
ax3.plot(tailles_np, temps_fusion_np / nlogn, 'd-', label='Fusion', linewidth=2)
ax3.set_xlabel('Taille N', fontsize=12)
ax3.set_ylabel('Temps / (N log N) (secondes)', fontsize=12)
ax3.set_title('Complexité O(n log n) - Temps divisé par N log N', fontsize=14, fontweight='bold')
ax3.legend()
ax3.grid(True, alpha=0.3)

# Graphique 4: T(n) / log n (pour recherche dichotomique)
ax4 = axes[1, 1]
logn = np.log2(tailles_np)
ax4.plot(tailles_np, temps_dicho_np / logn, 'o-', label='Recherche Dicho', linewidth=2, color='green')
ax4.set_xlabel('Taille N', fontsize=12)
ax4.set_ylabel('Temps / log N (secondes)', fontsize=12)
ax4.set_title('Complexité O(log n) - Temps divisé par log N', fontsize=14, fontweight='bold')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n=== Analyse des Complexités ===")
print("\nPour O(n²) - courbe horizontale indique complexité quadratique:")
print("  - Tri par Insertion: O(n²)")
print("  - Tri par Sélection: O(n²)")
print("  - Tri par Bulles: O(n²)")
print("\nPour O(n log n) - courbe horizontale indique complexité linéarithmique:")
print("  - Tri par Fusion: O(n log n)")
print("\nPour O(log n) - courbe horizontale indique complexité logarithmique:")
print("  - Recherche Dichotomique: O(log n)")
```
