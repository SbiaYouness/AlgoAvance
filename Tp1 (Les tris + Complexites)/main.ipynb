# TP 1: Algos de Tris + Complexité

### Les Algorithmes de Tris

```python
import random
import time
import timeit 

N=100
list1= [None]*N
def reinit():
    global list1
    for i in range(N):
        list1[i]=(random.randint(1,1546612))
reinit()
print(list1)
```

1. **Tri par insertion**


```python
def trierInsertion(list1,N):

    for i in range(1,N):
        temp=list1[i]
        j=i-1
        while(j>=0 and temp<list1[j]):
            list1[j+1]=list1[j]
            j-=1

        list1[j+1]=temp

print(list1[12])
```

**- Calcul de Complexité**

```python
moyenT=0

for k in range(10): #from 0 to 9
    reinit()

    debut = time.time()
    trierInsertion(list1,N)
    fin = time.time()

    moyenT+=fin-debut

print(f"Le temps moyen quand N={N} par insertion est: {moyenT/10}")
```

2. **Tri par selection**

```python
def trierSelection(list1,N):
    for i in range(N-1):
        min_indice=i
        for j in range(i+1,N):
            if list1[min_indice]>list1[j]:
                min_indice=j
        tempsel=list1[i]
        list1[i]=list1[min_indice]
        list1[min_indice]=tempsel

trierSelection(list1,N)
print(list1)
```

**- Calcul de Complexité**

```python
moyenT=0

for k in range(10): #from 0 to 9
    reinit()

    debut = time.time()
    trierSelection(list1,N)
    fin = time.time()
    
    moyenT+=fin-debut

print(f"Le temps moyen quand N={N} par selection est: {moyenT/10}")
```

3. **Tri par bulles**

```python
def trierBulles(list1,N):
    for i in range(N-1):
        for j in range(N-(i+1)):
            if list1[j]>list1[j+1]:
                tempsel=list1[j]
                list1[j]=list1[j+1]
                list1[j+1]=tempsel

trierBulles(list1,N)
print(list1)
```

**- Calcul de Complexité**

```python
moyenT=0

for k in range(10): #from 0 to 9
    reinit()

    debut = time.time()
    trierBulles(list1,N)
    fin = time.time()
    
    moyenT+=fin-debut

print(f"Le temps moyen quand N={N} par bulles est: {moyenT/10}")
```

4. **Recherche dichotomique**

```python
def rechdicho(list1,N,val):
    debut = 0
    fin = N - 1

    while debut <= fin:
        milieu = (debut + fin) // 2

        if list1[milieu] == val:
            return milieu  
        elif val < list1[milieu]:
            fin = milieu - 1
        else:
            debut = milieu + 1

    return -1
    
print(rechdicho(list1,N, list1[15]))
```

**- Calcul de Complexité**

```python
moyenT=0

for k in range(10): #from 0 to 9
    reinit()

    debut = time.time()
    rechdicho(list1,N, list1[15])
    fin = time.time()
    
    moyenT+=fin-debut

print(f"Le temps moyen quand N={N} recherche dicho est: {moyenT/10}")
```

5. **Tri par fusion**

```python
def tri_par_fusion(liste):
    if len(liste) <= 1:
        return liste
    
    milieu = len(liste) // 2
    gauche = tri_par_fusion(liste[:milieu])
    droite = tri_par_fusion(liste[milieu:])
    
    return fusion(gauche, droite)
    
def fusion(gauche, droite):
    resultat = []
    i = j = 0
    
    while i < len(gauche) and j < len(droite):
        if gauche[i] <= droite[j]:
            resultat.append(gauche[i])
            i += 1
        else:
            resultat.append(droite[j])
            j += 1
    
    resultat.extend(gauche[i:])
    resultat.extend(droite[j:])
    return resultat

# Test
liste_test = tri_par_fusion(list1.copy())
print("Liste triée par fusion:", liste_test[:10], "...")
```

**- Calcul de Complexité**

```python
moyenT=0

for k in range(10): #from 0 to 9
    reinit()

    debut = time.time()
    tri_par_fusion(list1)
    fin = time.time()
    
    moyenT+=fin-debut

print(f"Le temps moyen quand N={N} recherche fusion est: {moyenT/10}")
```

6. **Tri par Tas (Heap Sort)**

```python
def entasser(arr, n, i):
    maximum = i
    gauche = 2*i + 1
    droite = 2*i + 2
    
    if gauche < n and arr[gauche] > arr[maximum]:
        maximum = gauche
    if droite < n and arr[droite] > arr[maximum]:
        maximum = droite
    
    if maximum != i:
        arr[i], arr[maximum] = arr[maximum], arr[i]
        entasser(arr, n, maximum)

def trierTas(list1, N):
    # Construire le tas max
    for i in range(N//2 - 1, -1, -1):
        entasser(list1, N, i)
    
    # Extraire les éléments un par un
    for i in range(N-1, 0, -1):
        list1[i], list1[0] = list1[0], list1[i]
        entasser(list1, i, 0)

trierTas(list1, N)
print(list1)
```

**- Calcul de Complexité**

```python
moyenT=0

for k in range(10): #from 0 to 9
    reinit()

    debut = time.time()
    trierTas(list1, N)
    fin = time.time()
    
    moyenT+=fin-debut

print(f"Le temps moyen quand N={N} par tas est: {moyenT/10}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Différentes valeurs de N à tester
tailles = [100, 500, 1000, 5000, 10000, 50000]

# Stocker les temps moyens
temps_insertion = []
temps_selection = []
temps_bulles = []
temps_fusion = []
temps_tas = []
temps_dicho = []

# Mesurer les temps
for n in tailles:
    print(f"Testing N = {n}...")
    N_old = N
    N = n
    
    # Tri par insertion
    moyenT = 0
    for k in range(10):
        list1 = [random.randint(0, 1500000) for _ in range(N)]
        debut = time.time()
        trierInsertion(list1, N)
        fin = time.time()
        moyenT += fin - debut
    temps_insertion.append(moyenT / 10)
    
    # Tri par sélection
    moyenT = 0
    for k in range(10):
        reinit()
        debut = time.time()
        trierSelection(list1, N)
        fin = time.time()
        moyenT += fin - debut
    temps_selection.append(moyenT / 10)
    
    # Tri par bulles
    moyenT = 0
    for k in range(10):
        reinit()
        debut = time.time()
        trierBulles(list1, N)
        fin = time.time()
        moyenT += fin - debut
    temps_bulles.append(moyenT / 10)
    
    # Tri par fusion
    moyenT = 0
    for k in range(10):
        reinit()
        debut = time.time()
        tri_par_fusion(list1.copy())
        fin = time.time()
        moyenT += fin - debut
    temps_fusion.append(moyenT / 10)
    
    # Tri par tas
    moyenT = 0
    for k in range(10):
        reinit()
        debut = time.time()
        trierTas(list1, N)
        fin = time.time()
        moyenT += fin - debut
    temps_tas.append(moyenT / 10)
    
    # Recherche dichotomique
    moyenT = 0
    for k in range(10):
        reinit()
        trierInsertion(list1, N)
        val = list1[N // 2]
        debut = time.time()
        rechdicho(list1, N, val)
        fin = time.time()
        moyenT += fin - debut
    temps_dicho.append(moyenT / 10)
    
    N = N_old

print("Mesures terminées!")
```

```python
# Convertir en arrays
n = np.array(tailles)
t_insertion = np.array(temps_insertion)
t_selection = np.array(temps_selection)
t_bulles = np.array(temps_bulles)
t_fusion = np.array(temps_fusion)
t_tas = np.array(temps_tas)
t_dicho = np.array(temps_dicho)

# Calculer les normalisations
logn = np.log2(n)
nlogn = n * np.log2(n)

# Créer les graphiques - Un graphique par algorithme
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Graphique 1: Tri par Insertion
ax1 = axes[0, 0]
ax1.plot(n, t_insertion / logn, 'd-', label='T / log N', linewidth=2, color='purple')
ax1.plot(n, t_insertion / n, 'o-', label='T / N', linewidth=2, color='blue')
ax1.plot(n, t_insertion / (n * logn), 's-', label='T / (N log N)', linewidth=2, color='green')
ax1.plot(n, t_insertion / (n**2), '^-', label='T / N²', linewidth=2, color='red')
ax1.set_xlabel('Taille N', fontsize=11)
ax1.set_ylabel('Temps normalisé', fontsize=11)
ax1.set_title('Tri par Insertion', fontsize=13, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Graphique 2: Tri par Sélection
ax2 = axes[0, 1]
ax2.plot(n, t_selection / logn, 'd-', label='T / log N', linewidth=2, color='purple')
ax2.plot(n, t_selection / n, 'o-', label='T / N', linewidth=2, color='blue')
ax2.plot(n, t_selection / (n * logn), 's-', label='T / (N log N)', linewidth=2, color='green')
ax2.plot(n, t_selection / (n**2), '^-', label='T / N²', linewidth=2, color='red')
ax2.set_xlabel('Taille N', fontsize=11)
ax2.set_ylabel('Temps normalisé', fontsize=11)
ax2.set_title('Tri par Sélection', fontsize=13, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Graphique 3: Tri par Bulles
ax3 = axes[0, 2]
ax3.plot(n, t_bulles / logn, 'd-', label='T / log N', linewidth=2, color='purple')
ax3.plot(n, t_bulles / n, 'o-', label='T / N', linewidth=2, color='blue')
ax3.plot(n, t_bulles / (n * logn), 's-', label='T / (N log N)', linewidth=2, color='green')
ax3.plot(n, t_bulles / (n**2), '^-', label='T / N²', linewidth=2, color='red')
ax3.set_xlabel('Taille N', fontsize=11)
ax3.set_ylabel('Temps normalisé', fontsize=11)
ax3.set_title('Tri par Bulles', fontsize=13, fontweight='bold')
ax3.legend()
ax3.grid(True, alpha=0.3)

# Graphique 4: Tri par Fusion
ax4 = axes[1, 0]
ax4.plot(n, t_fusion / logn, 'd-', label='T / log N', linewidth=2, color='purple')
ax4.plot(n, t_fusion / n, 'o-', label='T / N', linewidth=2, color='blue')
ax4.plot(n, t_fusion / (n * logn), 's-', label='T / (N log N)', linewidth=2, color='green')
ax4.plot(n, t_fusion / (n**2), '^-', label='T / N²', linewidth=2, color='red')
ax4.set_xlabel('Taille N', fontsize=11)
ax4.set_ylabel('Temps normalisé', fontsize=11)
ax4.set_title('Tri par Fusion', fontsize=13, fontweight='bold')
ax4.legend()
ax4.grid(True, alpha=0.3)

# Graphique 5: Tri par Tas
ax5 = axes[1, 1]
ax5.plot(n, t_tas / logn, 'd-', label='T / log N', linewidth=2, color='purple')
ax5.plot(n, t_tas / n, 'o-', label='T / N', linewidth=2, color='blue')
ax5.plot(n, t_tas / (n * logn), 's-', label='T / (N log N)', linewidth=2, color='green')
ax5.plot(n, t_tas / (n**2), '^-', label='T / N²', linewidth=2, color='red')
ax5.set_xlabel('Taille N', fontsize=11)
ax5.set_ylabel('Temps normalisé', fontsize=11)
ax5.set_title('Tri par Tas (Heap Sort)', fontsize=13, fontweight='bold')
ax5.legend()
ax5.grid(True, alpha=0.3)

# Graphique 6: Recherche Dichotomique
ax6 = axes[1, 2]
ax6.plot(n, t_dicho / logn, 'd-', label='T / log N', linewidth=2, color='purple')
ax6.plot(n, t_dicho / n, 'o-', label='T / N', linewidth=2, color='blue')
ax6.plot(n, t_dicho / (n * logn), 's-', label='T / (N log N)', linewidth=2, color='green')
ax6.plot(n, t_dicho / (n**2), '^-', label='T / N²', linewidth=2, color='red')
ax6.set_xlabel('Taille N', fontsize=11)
ax6.set_ylabel('Temps normalisé', fontsize=11)
ax6.set_title('Recherche Dichotomique', fontsize=13, fontweight='bold')
ax6.legend()
ax6.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n=== Analyse de Complexité ===")
print("\n📊 Pour chaque algorithme, la courbe HORIZONTALE indique la complexité correcte:")
print("\n1. Tri par Insertion: O(n²) → courbe T/N² (rouge) horizontale")
print("2. Tri par Sélection: O(n²) → courbe T/N² (rouge) horizontale")
print("3. Tri par Bulles: O(n²) → courbe T/N² (rouge) horizontale")
print("4. Tri par Fusion: O(n log n) → courbe T/(N log N) (verte) horizontale")
print("5. Tri par Tas: O(n log n) → courbe T/(N log N) (verte) horizontale")
print("6. Recherche Dichotomique: O(log n) → courbe T/log N (violette) horizontale")
```
