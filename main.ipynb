# TP 1: Algos de Tris + Complexité

### Les Algorithmes de Tris

```python
import random
import time
import timeit 

N=100
list1= [None]*N
def reinit():
    global list1
    for i in range(N):
        list1[i]=(random.randint(1,1546612))
reinit()
print(list1)
```

1. **Tri par insertion**


```python
def trierInsertion(list1,N):

    for i in range(1,N):
        temp=list1[i]
        j=i-1
        while(j>=0 and temp<list1[j]):
            list1[j+1]=list1[j]
            j-=1

        list1[j+1]=temp

print(list1[12])
```

**- Calcul de Complexité**

```python
moyenT=0

for k in range(10): #from 0 to 9
    reinit()

    debut = time.time()
    trierInsertion(list1,N)
    fin = time.time()

    moyenT+=fin-debut

print(f"Le temps moyen quand N={N} par insertion est: {moyenT/10}")
```

2. **Tri par selection**

```python
def trierSelection(list1,N):
    for i in range(N-1):
        min_indice=i
        for j in range(i+1,N):
            if list1[min_indice]>list1[j]:
                min_indice=j
        tempsel=list1[i]
        list1[i]=list1[min_indice]
        list1[min_indice]=tempsel

trierSelection(list1,N)
print(list1)
```

**- Calcul de Complexité**

```python
moyenT=0

for k in range(10): #from 0 to 9
    reinit()

    debut = time.time()
    trierSelection(list1,N)
    fin = time.time()
    
    moyenT+=fin-debut

print(f"Le temps moyen quand N={N} par selection est: {moyenT/10}")
```

3. **Tri par bulles**

```python
def trierBulles(list1,N):
    for i in range(N-1):
        for j in range(N-(i+1)):
            if list1[j]>list1[j+1]:
                tempsel=list1[j]
                list1[j]=list1[j+1]
                list1[j+1]=tempsel

trierBulles(list1,N)
print(list1)
```

**- Calcul de Complexité**

```python
moyenT=0

for k in range(10): #from 0 to 9
    reinit()

    debut = time.time()
    trierBulles(list1,N)
    fin = time.time()
    
    moyenT+=fin-debut

print(f"Le temps moyen quand N={N} par bulles est: {moyenT/10}")
```

4. **Recherche dichotomique**

```python
def rechdicho(list1,N,val):
    debut = 0
    fin = N - 1

    while debut <= fin:
        milieu = (debut + fin) // 2

        if list1[milieu] == val:
            return milieu  
        elif val < list1[milieu]:
            fin = milieu - 1
        else:
            debut = milieu + 1

    return -1
    
print(rechdicho(list1,N, list1[15]))
```

**- Calcul de Complexité**

```python
moyenT=0

for k in range(10): #from 0 to 9
    reinit()

    debut = time.time()
    rechdicho(list1,N, list1[15])
    fin = time.time()
    
    moyenT+=fin-debut

print(f"Le temps moyen quand N={N} recherche dicho est: {moyenT/10}")
```

5. **Tri par fusion**

```python
def tri_par_fusion(liste):
    if len(liste) <= 1:
        return liste
    
    milieu = len(liste) // 2
    gauche = tri_par_fusion(liste[:milieu])
    droite = tri_par_fusion(liste[milieu:])
    
    return fusion(gauche, droite)
    
def fusion(gauche, droite):
    resultat = []
    i = j = 0
    
    while i < len(gauche) and j < len(droite):
        if gauche[i] <= droite[j]:
            resultat.append(gauche[i])
            i += 1
        else:
            resultat.append(droite[j])
            j += 1
    
    resultat.extend(gauche[i:])
    resultat.extend(droite[j:])
    return resultat

# Test
liste_test = tri_par_fusion(list1.copy())
print("Liste triée par fusion:", liste_test[:10], "...")
```

**- Calcul de Complexité**

```python
moyenT=0

for k in range(10): #from 0 to 9
    reinit()

    debut = time.time()
    tri_par_fusion(list1)
    fin = time.time()
    
    moyenT+=fin-debut

print(f"Le temps moyen quand N={N} recherche fusion est: {moyenT/10}")
```

6. **Tri par Tas (Heap Sort)**

```python
def entasser(arr, n, i):
    maximum = i
    gauche = 2*i + 1
    droite = 2*i + 2
    
    if gauche < n and arr[gauche] > arr[maximum]:
        maximum = gauche
    if droite < n and arr[droite] > arr[maximum]:
        maximum = droite
    
    if maximum != i:
        arr[i], arr[maximum] = arr[maximum], arr[i]
        entasser(arr, n, maximum)

def trierTas(list1, N):
    # Construire le tas max
    for i in range(N//2 - 1, -1, -1):
        entasser(list1, N, i)
    
    # Extraire les éléments un par un
    for i in range(N-1, 0, -1):
        list1[i], list1[0] = list1[0], list1[i]
        entasser(list1, i, 0)

trierTas(list1, N)
print(list1)
```

**- Calcul de Complexité**

```python
moyenT=0

for k in range(10): #from 0 to 9
    reinit()

    debut = time.time()
    trierTas(list1, N)
    fin = time.time()
    
    moyenT+=fin-debut

print(f"Le temps moyen quand N={N} par tas est: {moyenT/10}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Différentes valeurs de N à tester
tailles = [100, 1000, 10000, 100000]

# Stocker les temps moyens
temps_insertion = []
temps_selection = []
temps_bulles = []
temps_fusion = []
temps_tas = []
temps_dicho = []

# Mesurer les temps
for n in tailles:
    print(f"Testing N = {n}...")
    N_old = N
    N = n
    
    # Tri par insertion
    moyenT = 0
    for k in range(10):
        reinit()
        debut = time.time()
        trierInsertion(list1, N)
        fin = time.time()
        moyenT += fin - debut
    temps_insertion.append(moyenT / 10)
    
    # Tri par sélection
    moyenT = 0
    for k in range(10):
        reinit()
        debut = time.time()
        trierSelection(list1, N)
        fin = time.time()
        moyenT += fin - debut
    temps_selection.append(moyenT / 10)
    
    # Tri par bulles
    moyenT = 0
    for k in range(10):
        reinit()
        debut = time.time()
        trierBulles(list1, N)
        fin = time.time()
        moyenT += fin - debut
    temps_bulles.append(moyenT / 10)
    
    # Tri par fusion
    moyenT = 0
    for k in range(10):
        reinit()
        debut = time.time()
        tri_par_fusion(list1.copy())
        fin = time.time()
        moyenT += fin - debut
    temps_fusion.append(moyenT / 10)
    
    # Tri par tas
    moyenT = 0
    for k in range(10):
        reinit()
        debut = time.time()
        trierTas(list1, N)
        fin = time.time()
        moyenT += fin - debut
    temps_tas.append(moyenT / 10)
    
    # Recherche dichotomique
    moyenT = 0
    for k in range(10):
        reinit()
        trierInsertion(list1, N)
        val = list1[N // 2]
        debut = time.time()
        rechdicho(list1, N, val)
        fin = time.time()
        moyenT += fin - debut
    temps_dicho.append(moyenT / 10)
    
    N = N_old

print("Mesures terminées!")
```

```python
# Convertir en arrays
n = np.array(tailles)
t_insertion = np.array(temps_insertion)
t_selection = np.array(temps_selection)
t_bulles = np.array(temps_bulles)
t_fusion = np.array(temps_fusion)
t_tas = np.array(temps_tas)
t_dicho = np.array(temps_dicho)

# Créer les graphiques
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Graphique 1: Algorithmes O(n²) - Temps / n²
ax1 = axes[0]
ax1.plot(n, t_insertion / (n**2), 'o-', label='Insertion', linewidth=2)
ax1.plot(n, t_selection / (n**2), 's-', label='Sélection', linewidth=2)
ax1.plot(n, t_bulles / (n**2), '^-', label='Bulles', linewidth=2)
ax1.set_xlabel('Taille N', fontsize=12)
ax1.set_ylabel('Temps / N²', fontsize=12)
ax1.set_title('Algorithmes O(n²)', fontsize=14, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Graphique 2: Algorithmes O(n log n) - Temps / (n log n)
ax2 = axes[1]
nlogn = n * np.log2(n)
ax2.plot(n, t_fusion / nlogn, 'o-', label='Fusion', linewidth=2, color='green')
ax2.plot(n, t_tas / nlogn, 's-', label='Tas', linewidth=2, color='orange')
ax2.set_xlabel('Taille N', fontsize=12)
ax2.set_ylabel('Temps / (N log N)', fontsize=12)
ax2.set_title('Algorithmes O(n log n)', fontsize=14, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Graphique 3: Recherche O(log n) - Temps / log n
ax3 = axes[2]
logn = np.log2(n)
ax3.plot(n, t_dicho / logn, 'o-', label='Recherche Dicho', linewidth=2, color='red')
ax3.set_xlabel('Taille N', fontsize=12)
ax3.set_ylabel('Temps / log N', fontsize=12)
ax3.set_title('Recherche O(log n)', fontsize=14, fontweight='bold')
ax3.legend()
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n=== Analyse ===")
print("Courbe horizontale = complexité correcte")
print("- Insertion, Sélection, Bulles: O(n²)")
print("- Fusion, Tas: O(n log n)")
print("- Recherche Dichotomique: O(log n)")
```
